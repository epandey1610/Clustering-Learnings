```python
# now find some outliers in the dataset


for i in df.loc[:,num_list]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    up = Q3 + 1.5*IQR
    low = Q1 - 1.5*IQR

    if df[(df[i] > up) | (df[i] < low)].any(axis=None):
        print(i,"yes")
    else:
        print(i, "no")
```

```python
# remove outliers;
for i in df.loc[:,out_list]:

    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    up_lim=Q3+1.5 *IQR
    low_lim=Q1-1.5 *IQR

    df.loc[df[i]>up_lim,i]=up_lim
    df.loc[df[i]<low_lim,i]=low_lim
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn import metrics

# Clustering

# K-Means
km = KMeans(n_clusters=3)
km.fit(df)
labels_km = km.labels_

# Hierarchical clustering
ac = AgglomerativeClustering(n_clusters=3)
ac.fit(df)
labels_ac = ac.labels_

# DBSCAN
# Increase the eps parameter to allow for larger clusters
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan.fit(df)
labels_dbscan = dbscan.labels_

# Evaluate clustering results

# Silhouette score for K-Means
silhouette_score_km = metrics.silhouette_score(df, labels_km)
print(f"KMeans Silhouette Score: {silhouette_score_km}")

# Silhouette score for Hierarchical clustering
silhouette_score_ac = metrics.silhouette_score(df, labels_ac)
print(f"Hierarchical Silhouette Score: {silhouette_score_ac}")

# Check if DBSCAN assigned all points to a single cluster
if len(set(labels_dbscan)) > 1:
    silhouette_score_dbscan = metrics.silhouette_score(df, labels_dbscan)
    print(f"DBSCAN Silhouette Score: {silhouette_score_dbscan}")
else:
    print("DBSCAN assigned all points to a single cluster.")

# Experiment with different parameter settings

# Elbow curve for KMeans
inertia_scores = []
for k in range(2, 10):
    km = KMeans(n_clusters=k)
    km.fit(df)
    inertia_scores.append(km.inertia_)

plt.plot(range(2, 10), inertia_scores)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()








```

```python
#The results of the clustering analysis suggest that the data can be divided into three distinct clusters.

#K-Means clustering:  The silhouette score for K-Means clustering is 0.38, which indicates that the clusters are reasonably well-separated.
#The elbow curve suggests that the optimal number of clusters is 3.


#Hierarchical clustering: The silhouette score for hierarchical clustering is 0.36, which is slightly lower than the score for K-Means clustering.
#This suggests that hierarchical clustering may not be as effective as K-Means clustering for this dataset.


#DBSCAN clustering: DBSCAN clustering assigned all points to a single cluster.
#This suggests that the data points are not well-separated and that DBSCAN is not a suitable clustering algorithm for this dataset.

#Hence, the results of the clustering analysis suggest that the data can be divided into three distinct clusters. K-Means clustering appears to be the most effective clustering algorithm for this dataset.
```
